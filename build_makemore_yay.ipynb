```python
# store in a 2d array where first col is first character, second character is second col
# Entry in the array is how often does it follow this patten/ 
# basically a frquency array or 28x28 (inlcuding special start and end)
# We will use pytorch for it.

import torch
```

```python
N = torch.zeros((28,28), dtype=torch.int32)
# 26 chars + 2 start and end character
```

```python
chars = sorted(list(set(''.join(words))))
# a..z basically set will remove duplicates, list will add it in a list form and then sorted from our dataset
```

```python
stoi = {s:i for i,s in enumerate(chars)}
stoi['<S>'] = 26
stoi['<E>'] = 27
```

```python
itos = {i:s for s,i in stoi.items()}
itos
```

```python
# Make a 
b = {}
for w in words:
    # Add a special token start and end tokens
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        i1 = stoi[ch1]
        i2 = stoi[ch2]
        N[i1][i2] += 1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(N, cmap='Blues')
```

```python
plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(28):
    for j in range(28):
        chstr = itos[i] + itos[j]
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray")
        plt.text(j, i, N[i,j].item(), ha="center", va="top", color='gray')
plt.axis('off')

# item to go from torch to integer, N[i][j] is tensor.
```

```python
# problems in S, E rows, S cannot be a end character so a col of 0s, 
# same for E cannot be a starting character
# SE is not a word (start and end immedietly)
# fixing this by removing S and E completely for now, keeping . as a indicator of ends

N = torch.zeros((27,27), dtype=torch.int32)
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
itos
```

```python
# Make a 
b = {}
for w in words:
    # Add a special token start and end tokens
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        i1 = stoi[ch1]
        i2 = stoi[ch2]
        N[i1][i2] += 1
```

```python
plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(27):
    for j in range(27):
        chstr = itos[i] + itos[j]
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray")
        plt.text(j, i, N[i,j].item(), ha="center", va="top", color='gray')
plt.axis('off')
```

```python
#.a and a. are now both a valid row (for starting and ending)
# let's convert the frequency graph to a probablitiy graphj
p = N[0].float()
p = p / p.sum()
p
```

```python
g = torch.Generator().manual_seed(2147483647)
ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
ix
itos[ix]
```

```python
g = torch.Generator().manual_seed(2147483647)
for i in range(30):
    out = []
    ix = 0
    while True:
        p = N[ix].float()
        p = p / p.sum()
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
# preprocessing the probablities, add 1 for model smoothening to make divison by 0 to division by 1, you can add whatever 
P = (N+1).float()
# Use torch sum
# 0 is col dimension, 1 is row dimension
P.sum(1, keepdim=False)
P /= P.sum(1, keepdim=True)
```

```python
g = torch.Generator().manual_seed(2147483647)
for i in range(5):
    out = []
    ix = 0
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
log_likelyhood = 0.0
n= 0
for w in words:
    # Add a special token start and end tokens
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        i1 = stoi[ch1]
        i2 = stoi[ch2]
        prob = P[i1, i2]
        logprob = torch.log(prob)
        log_likelyhood += logprob
        n += 1
        #print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')
print(f'{log_likelyhood = }')
## Loss function lowest can be 0 since log can be 0 for full probability, and negative infinite for log(0)
nll = -log_likelyhood
print(f'{nll = }')
print(f'{nll/n = }')
```
